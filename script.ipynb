```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("CH_stim_SPK_data.csv",usecols=['Channel','stim_cond','SPK_count'])
df.head()
```

```python
df_ch_stimC = df.groupby(['Channel','stim_cond']).sum().reset_index().rename(columns={'SPK_count':'tot_SPK_count'})
df_ch_stimC
```

```python
fig, axs = plt.subplots(4,4, figsize=(20,20))
axs = axs.ravel() # flatten the axis array

for i,ch in zip(range(df_ch_stimC['Channel'].nunique()),df_ch_stimC['Channel'].unique()):
    
    #SPKcountMatrix.extend([np.array(df_ch_stimC[df_ch_stimC['Channel'] == ch]['tot_SPK_count']).reshape(2,2)])
    axs[i].imshow(np.array(df_ch_stimC[df_ch_stimC['Channel'] == ch]['tot_SPK_count']).reshape(2,2), cmap='Blues', interpolation='nearest')
    axs[i].set_title(df_ch_stimC['Channel'].unique()[i])

```
